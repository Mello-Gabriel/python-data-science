```python
sns.jointplot(x='fare',y='age',data=titanic)
```

```python
# PONHA O SEU CÓDIGO AQUI, 
# POIS SE USASSE A LINHA DEBAIXO, PODERIA ACABAR APAGANDO O PLOT QUE DESEJAMOS ENCONTRAR.
```

```python
sns.distplot(titanic['fare'],kde=False,color ='r',bins=30)
plt.xlim(0,600)
plt.ylim(0,500)
```

```python
# PONHA O SEU CÓDIGO AQUI, 
# POIS SE USASSE A LINHA DEBAIXO, PODERIA ACABAR APAGANDO O PLOT QUE DESEJAMOS ENCONTRAR.
sns.boxplot(x='class',y='age',data = titanic,palette='rainbow')
```

```python
# PONHA O SEU CÓDIGO AQUI, 
# POIS SE USASSE A LINHA DEBAIXO, PODERIA ACABAR APAGANDO O PLOT QUE DESEJAMOS ENCONTRAR.
sns.swarmplot(x='class',y='age',data=titanic,palette ='plasma')
```

```python
# PONHA O SEU CÓDIGO AQUI, 
# POIS SE USASSE A LINHA DEBAIXO, PODERIA ACABAR APAGANDO O PLOT QUE DESEJAMOS ENCONTRAR.
sns.countplot(titanic['sex'])
```

```python
# PONHA O SEU CÓDIGO AQUI, 
# POIS SE USASSE A LINHA DEBAIXO, PODERIA ACABAR APAGANDO O PLOT QUE DESEJAMOS ENCONTRAR.

correlacao = titanic.corr()
sns.heatmap(correlacao,cmap = 'PuBu')
plt.title('titanic.corr()')
```

```python
# PONHA O SEU CÓDIGO AQUI, 
# POIS SE USASSE A LINHA DEBAIXO, PODERIA ACABAR APAGANDO O PLOT QUE DESEJAMOS ENCONTRAR.
g = sns.FacetGrid(data=titanic, col='sex')
g.map(plt.hist, 'age')
```
